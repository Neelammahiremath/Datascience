```python

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('/content/Universities.csv')
```

```python
df1
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SATSCORES')
plt.ylabel('Frequence')
plt.title('Histogrma of SAT scores')
plt.show()
```

```python
sns.histplot(df1['SAT'])
plt.show()
```

```python
sns.histplot(df1['Accept'], color='blue', edgecolor='black')
plt.xlabel('Accept')
plt.ylabel('Frequency')
plt.title('Histogram of Accept')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Accept'],'bo')
plt.xlabel('SATscores')
plt.ylabel('ACCEPT')
plt.title('Scatter Plot of SAT vs ACCEPT')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Top10'],'rs')
plt.xlabel('SATscores')
plt.ylabel('Top 10 unver')
plt.title('Scatter Plot of SAT vs Top10')
plt.show()
```

```python
plt.scatter(df1['SAT'], df1['Top10'], c=df1['SAT'], cmap='rainbow')
plt.xlabel('SAT Scores')
plt.ylabel('Top 10 Percentage')
plt.title('Scatter Plot of SAT Scores vs Top 10 Percentage')
plt.colorbar(label='SAT Scores') # Add a colorbar to show the mapping
plt.show()
```

```python
plt.plot(df1['SAT'], df1['Top10'],'o')
plt.xlabel('SATScores')
plt.ylabel('Top 10')
plt.title('Scatter Plot of SAT Scores vs. Top 10')
plt.show()
```
